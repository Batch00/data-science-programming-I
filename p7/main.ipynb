```python
import csv

# Modified from https://automatetheboringstuff.com/chapter14/
def process_csv(filename):
    example_file = open(filename, encoding="utf-8")
    example_reader = csv.reader(example_file)
    example_data = list(example_reader)
    example_file.close()
    return example_data
```

```python
def get_month(date):
    """Returns the month when the date is the in the 'mm/dd/yyyy' format"""
    return int(date[:2])


def get_day(date):
    """Returns the day when the date is the in the 'mm/dd/yyyy' format"""
    return int(date[3:5])


def get_year(date):
    """Returns the year when the date is the in the 'mm/dd/yyyy' format"""
    return int(date[6:])

import datetime

def get_number_of_days(start_date, end_date):
    """Gets the number of days between the start_date (in 'mm/dd/yyyy' format) and end_date (in 'mm/dd/yyyy' format)"""
    day1 = datetime.datetime.strptime(start_date, '%m/%d/%Y') # The second argument is a format string to tell the function how to process the date string
    day2 = datetime.datetime.strptime(end_date, '%m/%d/%Y')
    delta = day2 - day1
    return delta.days
```

```python
#q10
def fully_vaccinated_by(date):
    '''return a dictionary mapping each country to the most recent number of people fully vaccinated by the given date; if no data is available, the value is None.'''
    fully_vax_dict = {}
    for row in covid_data:
        country_i = row[header.index('country')]
        date_i = row[header.index('date')]
        fully_vax_i = row[header.index('people_fully_vaccinated')]
        if country_i not in fully_vax_dict:
            if fully_vax_i == '':
                fully_vax_dict[country_i] = None
            else:
                fully_vax_dict[country_i] = int(fully_vax_i)
            #print(fully_vax_dict)
            #print(date_i)
        if get_number_of_days(date_i, date) >= 0:
            if fully_vax_i != '':
                fully_vax_dict[country_i] = int(fully_vax_i)
            #print(date_i)
    return fully_vax_dict

fully_vaccinated_by('10/16/2021')
```

```python
def cumulative_interpolation(column_name, date):
    '''return a dictionary mapping each country to the most recent number of people fully vaccinated by the given date; if no data is available, the value is None.'''
    cumulative_interpolation_dict = {}
    for row in covid_data:
        country_i = row[header.index('country')]
        date_i = row[header.index('date')]
        any_i = row[header.index(column_name)]
        if country_i not in cumulative_interpolation_dict:
            if any_i == '':
                cumulative_interpolation_dict[country_i] = None
            else:
                cumulative_interpolation_dict[country_i] = int(any_i)
            #print(fully_vax_dict)
            #print(date_i)
        if get_number_of_days(date_i, date) >= 0:
            if any_i != '':
                cumulative_interpolation_dict[country_i] = int(any_i)
            #print(date_i)
    return cumulative_interpolation_dict

cumulative_interpolation("daily_vaccinations", '10/12/2021')
```

```python
def ratio_doses(date):
    ratio_dict = {}
    for row in covid_data:
        country = row[header.index('country')]
        
        if country not in ratio_dict:
            fully = vaccination_stats[country][date]["people_fully_vaccinated"]
            at_least = vaccination_stats[country][date]["people_vaccinated"]
            if fully == None or at_least == None:
                ratio_dict[country] = None
            else:
                one_dose = at_least - fully
                ratio = one_dose / fully
                ratio_dict[country] = ratio
    return ratio_dict
                 
```

```python
#q18
min_val = 1
rat_dict = ratio_doses('10/16/2021')
for key in rat_dict:
    if rat_dict[key] == None:
        continue
    elif rat_dict[key] < min_val:
        min_val = rat_dict[key]
        min_val_key = key
min_val_key
```

```python
def vac_per_capita(date):
    vac_per_capita_dict = {}
    for row in covid_data:
        country = row[header.index('country')]
        if country not in vac_per_capita_dict:
            total = vaccination_stats[country][date]["total_vaccinations"]
            population = vaccination_stats[country][date]["population"]
            #print(population)
            if total != None:
                per_cap = total / population
                vac_per_capita_dict[country] = per_cap
                
    return vac_per_capita_dict
```

```python
def vac_per_pop(date):
    vac_per_pop_dict = {}
    for row in covid_data:
        country = row[header.index('country')]
        if country not in vac_per_pop_dict:
            total = vaccination_stats[country][date]["daily_vaccinations"]
            population = vaccination_stats[country][date]["population"]
            #print(population)
            if total != None:
                per_pop = total / population
                vac_per_pop_dict[country] = per_pop
                
    return vac_per_pop_dict
vac_per_pop('10/10/2021')
```
